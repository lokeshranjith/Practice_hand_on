```python
# Import requried library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("netflix_titles.csv")
```

```python
data.head()
```

```python
labels=['TV Show','Movie']
plt.pie(data['type'].value_counts().sort_values(),
       labels=labels,autopct="%1.2f%%",explode=(0,0.2),
       startangle=90,colors=['blue','red'])
plt.title("Type of Netflix Content")
plt.axis("equal")
plt.show()
```
