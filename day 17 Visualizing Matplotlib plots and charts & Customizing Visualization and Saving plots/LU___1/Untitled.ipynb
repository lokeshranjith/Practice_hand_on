```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
car=pd.read_csv("Car_sales.csv")
```

```python
car.head()
```

```python
car.columns
```

```python
man=car.groupby("Manufacturer").mean()
man.head(3)
```

```python
plt.barh(ma.index,car['4-year resale value'])
plt.ylabel("Manufacturer")
plt.xlabel("4-year resale value")
```

```python
man=car['Manufacturer']
resale=car["4-year resale value"]
plt.barh(man,resale)
```

```python
#Mean 4-year resale value for each Manufacturer
plt.figure(figsize=(10,8)) # Creates a new figure and allows you to set different parameters like figuresize()
for group,data in car.groupby('Manufacturer'):
    plt.barh(group,data['4-year resale value'].mean())
```
