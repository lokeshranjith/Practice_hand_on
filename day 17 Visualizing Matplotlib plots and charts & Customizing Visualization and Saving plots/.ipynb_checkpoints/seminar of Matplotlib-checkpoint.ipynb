```python
jan_s=rain['JAN'].sort_values(ascending=False)
feb_s=rain.FEB.sort_values()
```

```python
jan_s
```

```python
print(feb_s)
```

```python

plt.figure(figsize=(6,2))
plt.bar(rain.YEAR[-10:],jan_s[-10:],color='b',label="JAN")
plt.bar(rain.YEAR[-10:],feb_s[-10:],color="r",label="FEB")
plt.xlabel("Years")
plt.ylabel("Month of JAN and FEB")
plt.title("RAINFALL VISUAL")
plt.legend(title="Month")
plt.show()
```
