```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Bar plot

```python
time_data={"facebook":120,"youutube":90,"instagram":40,"snapchat":28}
social_media=list(time_data.keys())
time_spend=list(time_data.values())
```

```python
plt.figure(figsize=(5,5))
color=['red','blue','deeppink','yellow']
label=["Youtube","Facebook","instagram","Snapchat"]
plt.bar(social_media,time_spend,label=label, color=color,width=0.5)
plt.xlabel("Social_media")
plt.ylabel("time_spend")
plt.title("Bar graph")
plt.legend()
plt.show()
```

## Multiple bar chart

```python
fb=(120,39,90,20,29)
yt=(89,121,21,33,44)
ig=(23,40,20,20,23)
sc=(20,40,20,30,90)
```

```python
x=np.arange(len(fb))
y=np.arange(len(yt))
z=np.arange(len(ig))
z1=np.arange(len(sc))
```

```python
plt.figure(figsize=(10,7))
barwidth=0.2
plt.bar(x,fb,color="b",width=barwidth,label="Facebook")
plt.bar(x+0.2, yt,color="r",width=barwidth,label="Youtube")
plt.bar(x+0.4, ig,color="deeppink",width=barwidth,label="Instagram")
plt.bar(x+0.6, sc,color="yellow",width=barwidth,label="Snapchat")
plt.xticks(x+0.3,("Monday","tuesday","wednesday","thursday","Friday"))
plt.xlabel("Day")
plt.ylabel("Time spend in minutes")
plt.title("Time spending on Social media")
plt.legend()
plt.show()
```

## Stacked Bar chart

```python
fb=(120,39,90,20,29)
yt=(89,121,21,33,44)
xy=["monday","tuesday","wednesday","thursday","friday"]
x=np.arange(len(x))
plt.figure(figsize=(10,7))
plt.bar(x,fb,color="b",width=0.4,label="Facebook")
plt.bar(x,yt,bottom=fb,color="red",width=0.4,label="Youtube")
plt.bar(x,ig,bottom=yt,color="deeppink",width=0.4,label="Instagram")
plt.bar(x,sc,bottom=ig,color="black",width=0.4,label="SnapaChat")
plt.xlabel("Day")
plt.ylabel("Time spend in minutes")
plt.title("Time spend on social media")
plt.xticks(x,xy)
plt.legend()
plt.show()
```
