```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("SampleSuperstore.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x="Category",y="Sales",data=df,estimator=np.mean,hue = 'class')
```

```python
x="Category"
y="Sales"
plt.bar(x,y,data=df,color="red")
plt.title("Sales on each category",fontsize=30)
plt.show()
```
