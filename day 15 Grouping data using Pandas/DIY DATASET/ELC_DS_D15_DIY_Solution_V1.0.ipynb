**Q1. Problem Statement: Data Grouping - I**

Load the "big_bang_theory.csv" dataset provided on LMS into a DataFrame, group the data based on column "directed_by" using the groupby() function and the grouped DataFrame. Now also print all the episodes and the data associated with them, which belongs to the director - Mark Cendroski.



**Step-1:** Loading the dataset in a DataFrame.

```python
# importing pandas package
import pandas as pd
import numpy as np
	
# making data frame from csv file
df = pd.read_csv("/content/big_bang_theory_episodes.csv").fillna(np.nan)
display(df)
```

**Step-2:** Grouping the data based on director names.

```python
#Grouping data based on director names
result = df.groupby('directed_by')
display(result.first())
```

**Step-3:** Printing the rows related to director- Mark Cendrowski

```python
#Printing individual groups
result.get_group('Mark Cendrowski')
```

**Q2. Problem Statement: Data Grouping - II**

Load the "big_bang_theory.csv" dataset provided on LMS into a DataFrame; Group the data based on two columns, "directed_by" and "written_by," using the groupby() function and print the grouped DataFrame. Also, print the average viewers in the United States and the director and the writer's details.


**Step-1:** Loading the dataset in a DataFrame.

```python
# importing pandas package
import pandas as pd
import numpy as np
	
# making data frame from csv file
df = pd.read_csv("/content/big_bang_theory_episodes.csv")
display(df)
```

**Step-2:** Average US viewers of each writer and director.

```python
#Grouping data based on director and writer
result = df.groupby(['directed_by', 'written_by']) ['us_viewers'].mean()
result
```
